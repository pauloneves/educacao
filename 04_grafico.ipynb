```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
plt.style.use('seaborn-whitegrid')
%matplotlib inline
```

```python
df = pd.read_feather('dados/final.feather')
df = df.iloc[::-1, :45]
df['rotulo'] = df['rank'].astype(int).astype(str) + '. ' +df.rotulo.str.split('/').str.get(0) 
```

```python
plt.rcParams["figure.figsize"] = (10,15)
plt.plot('mediana', 'rotulo', data=df, marker='d', linestyle='', alpha=.9)
plt.box(False)
plt.grid(False, axis='y')
```
